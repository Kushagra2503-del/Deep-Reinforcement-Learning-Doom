```python

import gymnasium
import imageio
import numpy as np
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv
from vizdoom import gymnasium_wrapper
from IPython.display import Video




def make_eval_env():

    return gymnasium.make("VizdoomBasic-v1", render_mode="rgb_array")


model = PPO.load("vizdoom_demo_agent")
env = DummyVecEnv([make_eval_env])


obs = env.reset()
images = []

print(" Recording Gameplay...")

for i in range(300):

    img = env.envs[0].render()
    images.append(img)


    action, _ = model.predict(obs)
    obs, _, done, _ = env.step(action)

    if done:
        break

env.close()


video_filename = "doom_final_cut.mp4"
imageio.mimsave(video_filename, np.array(images), fps=30)

print(f"Video saved: {video_filename}")
Video(video_filename, embed=True, width=640)
```
